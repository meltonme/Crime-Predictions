```python
import pandas as pd
#need zipfile module to unpack the zipfile through jupyter notebook
#module documentation https://pypi.org/project/zipfile36/
#https://stackoverflow.com/questions/26942476/reading-csv-zipped-files-in-python
import zipfile


#Upload the zipfile that we want to read
zf = zipfile.ZipFile('C:/Users/mahan/OneDrive/Desktop/Repos/Crime-Predictions/Resources/CMPD_Incidents.zip')

#Have Pandas read the csv file in the zip file 
CMPD_df = pd.read_csv(zf.open('CMPD_Incidents.csv'))

CMPD_df.head(5)
```

```python
#Change the values in the 'CLEARANCE STATUS' to 0 or 1
# 0 = Open
# 1 = Cleared  
#Define the mapping 
mapping = {'Open': 0, 'Cleared by Arrest': 1, 'Cleared by Arrest by Another Agency': 1, 'Exceptionally Cleared': 1, 'Unfounded': 1}

# Apply mapping to the 'CLEARANCE_STATUS' column and fill NaN values with a default value (e.g., False)
CMPD_df['CLEARANCE_STATUS'] = CMPD_df['CLEARANCE_STATUS'].map(mapping).fillna(False)

#check to make sure the above worked
CMPD_df.head(5)

#Convert the 'CLEARANCE_STATUS' column into an array 

status_array = CMPD_df['CLEARANCE_STATUS'].values
print(status_array)
```

```python
#Show the DataFrame with the Adjusted 'CLEARANCE STATUS'
CMPD_df.head(5)
```

```python
# List of violent and non-violent crimes based on the previous categorization
violent_crimes = [
    'Aggravated Assault', 'Affray', 'Animal Cruelty', 'Forcible Fondling', 
    'Forcible Rape', 'Forcible Sodomy', 'Human Trafficking, Commercial Sex Acts', 
    'Human Trafficking, Involuntary Servitude', 'Incest', 'Intimidation', 'Justifiable Homicide', 
    'Kidnapping', 'Murder', 'Negligent Manslaughter', 'Robbery', 'Sexual Assault With Object', 'Simple Assault', 
    'Statutory Rape'
]

non_violent_crimes = [
    'All Other Offenses', 'All Other Thefts', 'Assisting Gambling', 'Assisting Prostitution', 'Betting/Wagering', 
    'Bribery', 'Burglary/B&E', 'Counterfeiting/Forgery', 'Credit Card/Teller Fraud', 'Curfew/Loitering/Vagrancy Violations', 
    'Damage/Vandalism Of Property', 'Disorderly Conduct', 'Dog Bite/Animal Control Incident', 'Driving Under The Influence', 
    'Drug Equipment Violations', 'Drug/Narcotic Violations', 'Embezzlement', 'Extortion/Blackmail', 'False Pretenses/Swindle', 
    'Family Offenses; Nonviolent', 'Fire (Accidental/Non-Arson)', 'Gambling Equipment Violations', 'Gas Leak', 
    'Hacking/Computer Invasion', 'Identity Theft', 'Impersonation', 'Indecent Exposure', 'Liquor Law Violations', 
    'Missing Person', 'Motor Vehicle Theft', 'Other Unlisted Non-Criminal', 'Overdose', 'Peeping Tom', 
    'Pocket-Picking', 'Pornography/Obscene Material', 'Prostitution', 'Public Accident', 'Purchasing Prostitution', 
    'Purse-Snatching', 'Shoplifting', 'Stolen Property Offenses', 'Sudden/Natural Death Investigation', 'Suicide', 
    'Theft From Building', 'Theft From Coin-Operated Machine Or Device', 'Theft From Motor Vehicle', 
    'Theft of Motor Vehicle Parts from Vehicle', 'Trespass Of Real Property', 'Vehicle Recovery', 'Weapon Law Violations', 
    'Welfare Fraud', 'Wire Fraud', 'Worthless Check: Felony (over $2000)'
]

# Function to categorize the crimes
def categorize_crime(crime):
    if crime in violent_crimes:
        return 'Violent Crime'
    elif crime in non_violent_crimes:
        return 'Non-Violent Crime'
    else:
        return 'Unknown'

# Add a new column categorizing the values in the 'HIGHEST_NIBRS_DESCRIPTION' column
CMPD_df['Crime Category'] = CMPD_df['HIGHEST_NIBRS_DESCRIPTION'].apply(categorize_crime)

# Display the resulting DataFrame
CMPD_df
```

```python
#Delete the columns that we do not need
CMPD_df = CMPD_df.drop(columns=['X','Y','X_COORD_PUBLIC', 'Y_COORD_PUBLIC','NPA', 'CMPD_PATROL_DIVISION', 'GlobalID', 'DATE_INCIDENT_BEGAN', 'DATE_INCIDENT_END'])

#print to make sure that the columns actually dropped in the DataFrame
CMPD_df.columns
```

```python
# Print the original DataFrame
print("Original DataFrame:")
#print(CMPD_df)

# Remove rows with null values in column 'ZIP'
CMPD_df_filtered = CMPD_df.dropna(subset=['ZIP'])
print("\nDataFrame after removing rows with null values in column ZIP':")
CMPD_df_filtered
```

```python
#turn the DataFrame into a csv file

# Specify the file path where you want to save the CSV file
file_path = "data_CMPD.csv"

# Save the DataFrame to a CSV file without the index and with the header
CMPD_df_filtered.to_csv(file_path, index=False, header=True)
```
